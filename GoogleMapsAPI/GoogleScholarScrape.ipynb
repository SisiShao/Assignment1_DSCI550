```python

pg = ProxyGenerator()
pg.FreeProxies()
scholarly.use_proxy(pg)
```

```python
notfound = {}
f = {}
```

```python

for i,auth in enumerate(auths):
    try:
        if auth in f or auth in notfound or i==0:
            continue
        # Retrieve the author's data
        search_query = scholarly.search_author(auth)
        author = scholarly.fill(next(search_query))
                     
        # Get the author's publications
        publications = [pub for pub in author['publications']]
       
        f[auth] = [author['name'],author['affiliation'], publications, len(publications)]
        # found.append(f)

        if i==2:
            break
    except:
        notfound[auth] = 'N/A'
        continue


with open('gscholarNotFoundAlvin.json', 'w') as outj:
    json.dump(notfound, outj)

with open('gscholarFoundAlvin.json', 'w') as outj:
    json.dump(f, outj)
```
