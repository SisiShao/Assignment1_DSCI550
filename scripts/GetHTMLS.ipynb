```python
url = 'http://doi.org/'
skipped = []
try:
    for index,row in df.iterrows():
        if index < latest:
            continue
        title = row["Title"]
        for specialChar in '/\\:*?<>|"':
            title = title.replace(specialChar,"")
        title = title[:100]

        
        r = requests.get(url + row['DOI'])
        if r.status_code != 200:
            skipped.append(f'{r.status_code} skipped {url + row["DOI"]} {title}.html')
            continue

        html = r.content
        with open(f'../htmls/{title}.html','wb') as file:
            file.write(html)
except Exception as e:
    print(index,title)
    raise e
finally:
    latest = index
```

```python
with open('skipped.txt','w') as file:
    file.write('\n'.join(skipped))
```
