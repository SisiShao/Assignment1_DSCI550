```python
#function for convert all the 
def strip_accents(text):

    try:
        text = unicode(text, 'utf-8')
    except NameError: # unicode is a default on python 3 
        pass

    text = unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")

    return str(text)
```

```python
all_univrsity_info=[]
```

```python
def get_entries():
    url = "https://www.topuniversities.com/sites/default/files/qs-rankings-data/en/3740566_indicators.txt?1646373775?v=1646379988369"
    headers = {
        "user-agent": "Malfoy/5.0",
        "x-requested-with": "XMLHttpRequest"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    def make_pretty(entry):
        from bs4 import BeautifulSoup as Soup
        return {
            "Name": strip_accents(Soup(entry["uni"], "html.parser").select_one(".uni-link").get_text(strip=True)),
            "Rank": entry["overall_rank"],
            "Overall Score":Soup(entry["ind_18"], "html.parser").select_one(".td-wrap-in").get_text(strip=True),
            "Academic Reputation Score": Soup(entry["ind_76"], "html.parser").select_one(".td-wrap-in").get_text(strip=True),
            "Citations per Faculty": Soup(entry["ind_73"], "html.parser").select_one(".td-wrap-in").get_text(strip=True),
             "Faculty Student Ratio": Soup(entry["ind_36"], "html.parser").select_one(".td-wrap-in").get_text(strip=True)     
        }

    yield from map(make_pretty, response.json()["data"])
```

```python
university_result_list=[]
k=0
for i in range(len(df_list)):
    if k!=(len(list(get_entries()))-1):
        for entry in get_entries():
            if  re.match(df_list[i], entry['Name']):
                university_result_list.append(entry)
                k=0
                break
            else:
                k=k+1
    if k==(len(list(get_entries()))):
        university_result_list.append({'Name': df_list[i] , 'Rank': 'NaN', 'Overall Score': 'NaN', 'Academic Reputation Score': 'NaN', 'Citations per Faculty': 'NaN', 'Faculty Student Ratio': 'NaN'})
        k=0
```

```python
for i in range(len(university_result_list)):
    print(university_result_list[i])
```

```python
for i in range(len(university_result_list)):
    all_univrsity_info.append(university_result_list[i])
```
